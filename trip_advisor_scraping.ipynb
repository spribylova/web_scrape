```python
# search for the restaurant name on the Tripadvisor Homepage and copy paste url from webpage
# see that there are more Search Result Pages for the Most Recent Algorithm in english language, but much more reviews in other languages
# copy paste each SRP url and save html content to the soup object

url1 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
url2 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or10-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
url3 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or20-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
url4 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or30-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
url5 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or40-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
url6 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or50-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
# ...
url7 = 'https://www.tripadvisor.at/Restaurant_Review-g190445-d1093430-Reviews-or60-Lichtblick-Innsbruck_Tirol_Austrian_Alps.html'
# ...

headers={"User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.5042.108 Safari/537.36"}


response1 = requests.get(url1, headers=headers)
response2 = requests.get(url2, headers=headers)
response3 = requests.get(url3, headers=headers)
response4 = requests.get(url4, headers=headers)
response5 = requests.get(url5, headers=headers)
response6 = requests.get(url6, headers=headers)
# ...
response7 = requests.get(url7, headers=headers)
# ...

html_content1 = response1.content
html_content2 = response2.content
html_content3 = response3.content
html_content4 = response4.content
html_content5 = response5.content
html_content6 = response6.content
# ...
html_content7 = response7.content
# ...

soup1 = BeautifulSoup(response1.content, 'html.parser')
soup2 = BeautifulSoup(response2.content, 'html.parser')
soup3 = BeautifulSoup(response3.content, 'html.parser')
soup4 = BeautifulSoup(response4.content, 'html.parser')
soup5 = BeautifulSoup(response5.content, 'html.parser')
soup6 = BeautifulSoup(response6.content, 'html.parser')
# ...
soup7 = BeautifulSoup(response7.content, 'html.parser')
# ...
```

```python
# see how does the html code look like
# soup4
```

```python
reviews1 = []

for review1 in soup1.find_all('div', {'class': 'review-container'}):
    review_text1 = review1.find('p', {'class': 'partial_entry'}).text
    review_rating1 = review1.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews1.append((review_text1,review_rating1))

reviews2 = []

for review2 in soup2.find_all('div', {'class': 'review-container'}):
    review_text2 = review2.find('p', {'class': 'partial_entry'}).text
    review_rating2 = review2.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews2.append((review_text2,review_rating2))

reviews3 = []

for review3 in soup3.find_all('div', {'class': 'review-container'}):
    review_text3 = review3.find('p', {'class': 'partial_entry'}).text
    review_rating3 = review3.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews3.append((review_text3,review_rating3))

reviews4 = []

for review4 in soup4.find_all('div', {'class': 'review-container'}):
    review_text4 = review4.find('p', {'class': 'partial_entry'}).text
    review_rating4 = review4.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews4.append((review_text4,review_rating4))

reviews5 = []

for review5 in soup5.find_all('div', {'class': 'review-container'}):
    review_text5 = review5.find('p', {'class': 'partial_entry'}).text
    review_rating5 = review5.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews5.append((review_text5,review_rating5))

reviews6 = []

for review6 in soup6.find_all('div', {'class': 'review-container'}):
    review_text6 = review6.find('p', {'class': 'partial_entry'}).text
    review_rating6 = review6.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews6.append((review_text6,review_rating6))

reviews7 = []

for review7 in soup7.find_all('div', {'class': 'review-container'}):
    review_text7 = review7.find('p', {'class': 'partial_entry'}).text
    review_rating7 = review7.find('span', {'class': 'ui_bubble_rating'})['class'][1][1]
    reviews7.append((review_text7,review_rating7))
```

```python
for review1 in reviews1:
    review_text1 = review1[0].replace('\n', '').strip()
for review2 in reviews2:
    review_text2 = review2[0].replace('\n', '').strip()
for review3 in reviews3:
    review_text3 = review3[0].replace('\n', '').strip()
for review4 in reviews4:
    review_text4 = review4[0].replace('\n', '').strip()
for review5 in reviews5:
    review_text5 = review5[0].replace('\n', '').strip()
for review6 in reviews6:
    review_text6 = review6[0].replace('\n', '').strip()
for review7 in reviews7:
    review_text7 = review7[0].replace('\n', '').strip()
```

```python
# export the results to csv file
import csv

with open('reviews1.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review1 in reviews1:
        review_text1 = review1[0].replace('\n', '').strip()
        writer.writerow([review_text1])

with open('reviews2.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review2 in reviews2:
        review_text2 = review2[0].replace('\n', '').strip()
        writer.writerow([review_text2])

with open('reviews3.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review3 in reviews3:
        review_text3 = review3[0].replace('\n', '').strip()
        writer.writerow([review_text3])

with open('reviews4.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review4 in reviews4:
        review_text4 = review4[0].replace('\n', '').strip()
        writer.writerow([review_text4])

with open('reviews5.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review5 in reviews5:
        review_text5 = review5[0].replace('\n', '').strip()
        writer.writerow([review_text5])

with open('reviews6.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review6 in reviews6:
        review_text6 = review6[0].replace('\n', '').strip()
        writer.writerow([review_text6])

with open('reviews7.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Review Text'])
    for review7 in reviews7:
        review_text7 = review7[0].replace('\n', '').strip()
        writer.writerow([review_text7])
```
